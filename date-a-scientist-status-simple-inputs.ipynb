Lets keep it simple and only use 'job', 'age' and 'sex' to predict 'status'.

```python
data_select = data[['job', 'age', 'sex', 'status']]
data_select
```

```python
data_select = pd.get_dummies(
    data_select,
    columns = ['job', 'sex','status'],
    dtype = int
)
```

```python
data_select
```

```python
#One-hot encoding the columns with object data types
X_col_list = [col for col in data_select.columns if col not in ('status_available', 'status_married', 'status_seeing someone', 'status_single')]
y_col_list = ['status_available', 'status_married', 'status_seeing someone', 'status_single']
#y_col_list = ['status']

X_col_list
data_select
```

```python
#Creating tensors
import torch
X_tensor = torch.tensor(data_select[X_col_list].values, dtype = torch.float)
y_tensor = torch.tensor(data_select[y_col_list].values, dtype = torch.float)
```

```python
#Splitting data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_tensor, y_tensor, test_size = 0.2)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
from torch import nn
model = nn.Sequential(
    nn.Linear(24,14),
    nn.ReLU(),
    nn.Linear(14,4)
)
```

```python
import torch.optim as optim
from sklearn.metrics import accuracy_score, recall_score, precision_score

loss = nn.CrossEntropyLoss()
optimizer = optim.SGD(model.parameters(), lr= 0.3)

num_epochs = 1000
for epoch in range(num_epochs):
    predictions = model(X_train)
    CELoss = loss(predictions, y_train)
    CELoss.backward()
    optimizer.step()
    optimizer.zero_grad()
     
    if (epoch + 1) % 100 == 0:
        predicted_labels = torch.argmax(predictions, dim=1)
        true_labels = torch.argmax(y_train, dim=1)
        accuracy = accuracy_score(true_labels.cpu(), predicted_labels.cpu())
        recall = recall_score(true_labels.cpu(), predicted_labels.cpu(), average = 'weighted')
        #precision = precision_score(true_labels.cpu(), predicted_labels.cpu(), average = 'weighted')
        print(f'Epoch [{epoch+1}/{num_epochs}], CELoss: {CELoss.item():.4f}, Accuracy: {accuracy:.4f}, Recall: {recall:.4f}')

```
